```python
import requests
import json
```

```python
api_key = "IczooXKm4Ypx1whmf7sdn7CvcggKMEHbsrOo4UPz"  # Replace with your actual key
start_date = "2024-06-20"
end_date = "2024-07-25"

base_url = "https://api.nasa.gov/DONKI/"
endpoint = "CMEAnalysis"

# Construct the full URL with parameters
url = f"{base_url}{endpoint}?startDate={start_date}&endDate={end_date}&api_key={api_key}"
```

```python
try:
    response = requests.get(url)
    response.raise_for_status()  # This will raise an HTTPError for bad responses (4xx or 5xx)
    data = response.json()
    
    # Now you can work with your data in a Jupyter Notebook!
    print(json.dumps(data, indent=2))
    
except requests.exceptions.RequestException as e:
    print(f"Error making API request: {e}")
```

```python
if data:
    for event in data:
        print(f"Event Time: {event.get('startTime')}")
        print(f"Speed: {event.get('speed')}")
        print(f"Half-Angle: {event.get('halfAngle')}")
        print("-" * 20)
```
